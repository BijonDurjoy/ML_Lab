```python
import numpy as np
import matplotlib.pyplot as plt
```

Define the objective function

```python
def objective_function(x):
    return x**2 + 2*x + 1
```

Define the derivative of the objective function

```python
def gradient(x):
    return 2*x + 2
```

Gradient Descent function

```python
def gradient_descent(initial_x, learning_rate, num_iterations):
    x_values = []
    y_values = []

    x = initial_x

    for _ in range(num_iterations):
        x_values.append(x)
        y_values.append(objective_function(x))

        gradient_value = gradient(x)
        x = x - learning_rate * gradient_value

    return x_values, y_values
```

Set the initial values

```python
initial_x = -5
learning_rate = 0.1
num_iterations = 50
```

Run Gradient Descent

```python
x_values, y_values = gradient_descent(initial_x, learning_rate, num_iterations)
```

Plot the objective function and the gradient descent path

```python
x_range = np.linspace(-6, 2, 100)
plt.plot(x_range, objective_function(x_range), label='Objective Function')
plt.scatter(x_values, y_values, color='red', label='Gradient Descent Path')
plt.title('Gradient Descent Optimization')
plt.xlabel('X')
plt.ylabel('Objective Function Value')
plt.legend()
plt.show()
```
